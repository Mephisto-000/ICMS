# Two-Body-Problem Elliptical Orbit Demo

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
main_path = os.getcwd()
result_path = main_path + os.sep + "result.txt"
```

```python
rt = np.loadtxt(result_path, delimiter=",")
rx, ry = [], []
vx, vy = [], []

for i in rt:
    rx.append(i[1])
    ry.append(i[2])
    vx.append(i[3])
    vy.append(i[4])

for i, j, m, n in zip(rx, ry, vx, vy):
    print(i, j, m, n)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 10))
plt.plot(0.0, 0.0, color="blue", marker="o", markersize=20)
plt.plot(rx, ry, "r.")
plt.title("r(t)", fontsize=20)
plt.axis("equal")
plt.show()
```

```python
x_len = [i for i in range(len(rx))]
y_len = [i for i in range(len(ry))]
plt.figure(figsize=(20, 8))
plt.plot(x_len, rx, "r-o", label="rx")
plt.plot(y_len, ry, "b-o", label="ry")
plt.title("r(t), x and y positions", fontsize=20)
plt.legend(fontsize=15)
plt.show()
```
