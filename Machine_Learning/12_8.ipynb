# Example 12_8

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

## $x^{2}, x(0) = 1$

```python
x1_1 = tf.Variable(initial_value=1, name='x1', trainable=True, dtype=tf.float32)
x1_2 = tf.Variable(initial_value=1, name='x2', trainable=True, dtype=tf.float32)
x1_3 = tf.Variable(initial_value=1, name='x3', trainable=True, dtype=tf.float32)
x1_4 = tf.Variable(initial_value=1, name='x4', trainable=True, dtype=tf.float32)
```

```python
def x1_square():
    return x1_1**2
def x2_square():
    return x1_2**2
def x3_square():
    return x1_3**2
def x4_square():
    return x1_4**2
```

```python
optimizer1 = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=False)
optimizer2 = tf.keras.optimizers.RMSprop(learning_rate=0.01, rho=0.9, epsilon=None)
optimizer3 = tf.keras.optimizers.Adagrad(learning_rate=0.01, initial_accumulator_value=0.1, epsilon=1e-07)
optimizer4 = tf.keras.optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999, epsilon=1e-07)
```

```python
for i in range(100):
    optimizer1.minimize(loss=x1_square, var_list=[x1_1])

print(x1_1)
```

```python
for i in range(100):
    optimizer2.minimize(loss=x2_square, var_list=[x1_2])

print(x1_2)
```

```python
for i in range(100):
    optimizer3.minimize(loss=x3_square, var_list=[x1_3])

print(x1_3)
```

```python
for i in range(100):
    optimizer4.minimize(loss=x4_square, var_list=[x1_4])

print(x1_4)
```

```python
plt.figure(figsize=(12, 8))
x_plot_data = np.linspace(-1, 1, 100)
y_data = x_plot_data**2
plt.plot(x_plot_data, y_data, 'k-')
plt.scatter(x1_1, x1_1**2, color="red", s=70, label="SGD")
plt.scatter(x1_2, x1_2**2, color="brown", s=70, label="RMSprop")
plt.scatter(x1_3, x1_3**2, color="green", s=70, label="Adagrad")
plt.scatter(x1_4, x1_4**2, color="blue", s=70, label="Adam")
plt.legend(fontsize=20)
plt.show()
```

## $f(x) = x^{3} + sin(10x), x(0)=0.2$

```python
x2_1 = tf.Variable(initial_value=0.2, name='x1', trainable=True, dtype=tf.float32)
x2_2 = tf.Variable(initial_value=0.2, name='x2', trainable=True, dtype=tf.float32)
x2_3 = tf.Variable(initial_value=0.2, name='x3', trainable=True, dtype=tf.float32)
x2_4 = tf.Variable(initial_value=0.2, name='x4', trainable=True, dtype=tf.float32)
```

```python
def x1_f():
    return x2_1**3 + np.sin(10 * x2_1)
def x2_f():
    return x2_2**3 + np.sin(10 * x2_2)
def x3_f():
    return x2_3**3 + np.sin(10 * x2_3)
def x4_f():
    return x2_4**3 + np.sin(10 * x2_4)
```

```python
for i in range(50):
    optimizer1.minimize(loss=x1_f, var_list=[x2_1])

print(x2_1)
```

```python
for i in range(50):
    optimizer2.minimize(loss=x2_f, var_list=[x2_2])

print(x2_2)
```

```python
for i in range(50):
    optimizer3.minimize(loss=x3_f, var_list=[x2_3])

print(x2_3)
```

```python
for i in range(50):
    optimizer4.minimize(loss=x4_f, var_list=[x2_4])

print(x2_4)
```

```python
plt.figure(figsize=(12, 8))
x_plot_data = np.linspace(-1, 1, 100)
y_data = x_plot_data**3 + np.sin(10 * x_plot_data)
plt.plot(x_plot_data, y_data, 'k-')
plt.scatter(x2_1, x2_1**3 + np.sin(10 * x2_1), color="red", s=70, label="SGD")
plt.scatter(x2_2, x2_2**3 + np.sin(10 * x2_2), color="brown", s=70, label="RMSprop")
plt.scatter(x2_3, x2_3**3 + np.sin(10 * x2_3), color="green", s=70, label="Adagrad")
plt.scatter(x2_4, x2_4**3 + np.sin(10 * x2_4), color="blue", s=70, label="Adam")
plt.legend(fontsize=20)
plt.show()
```

## $f(x,y) = 0.5x^{2} + 2.5y^{2}, x(0) = 2.5, y(0)=2.5$

```python
x3_1 = tf.Variable(initial_value=2.5, name='x1', trainable=True, dtype=tf.float32)
x3_2 = tf.Variable(initial_value=2.5, name='x2', trainable=True, dtype=tf.float32)
x3_3 = tf.Variable(initial_value=2.5, name='x3', trainable=True, dtype=tf.float32)
x3_4 = tf.Variable(initial_value=2.5, name='x4', trainable=True, dtype=tf.float32)

y3_1 = tf.Variable(initial_value=2.5, name='y1', trainable=True, dtype=tf.float32)
y3_2 = tf.Variable(initial_value=2.5, name='y2', trainable=True, dtype=tf.float32)
y3_3 = tf.Variable(initial_value=2.5, name='y3', trainable=True, dtype=tf.float32)
y3_4 = tf.Variable(initial_value=2.5, name='y4', trainable=True, dtype=tf.float32)
```

```python
def xy1_f():
    return 0.5*x3_1**2 + 2.5*y3_1**2
def xy2_f():
    return 0.5*x3_2**2 + 2.5*y3_2**2
def xy3_f():
    return 0.5*x3_3**2 + 2.5*y3_3**2
def xy4_f():
    return 0.5*x3_4**2 + 2.5*y3_4**2
```

```python
for i in range(100):
    optimizer1.minimize(loss=xy1_f, var_list=[x3_1, y3_1])

print(x3_1, y3_1)
```

```python
for i in range(100):
    optimizer2.minimize(loss=xy2_f, var_list=[x3_2, y3_2])

print(x3_2, y3_2)
```

```python
for i in range(100):
    optimizer3.minimize(loss=xy3_f, var_list=[x3_3, y3_3])

print(x3_3, y3_3)
```

```python
for i in range(100):
    optimizer4.minimize(loss=xy4_f, var_list=[x3_4, y3_4])

print(x3_4, y3_4)
```

```python
fig = plt.figure(figsize=(18, 12))
ax = plt.axes(projection='3d')

x_plot_data = np.linspace(-2, 2, 100)
y_plot_data = np.linspace(-2, 2, 100)
def zf(x, y):
    return 0.5*x**2 + 2.5*y**2

X, Y = np.meshgrid(x_plot_data, y_plot_data)    
z_data = zf(X, Y)


ax.plot_surface(X, Y, z_data, rstride=1, cstride=1, cmap='binary')
ax.scatter3D(x3_1, y3_1, 0.5*x3_1**2 + 2.5*y3_1**2, color='red', s=90, label="SGD")
ax.scatter3D(x3_2, y3_2, 0.5*x3_2**2 + 2.5*y3_2**2, color='brown', s=90, label="RMSprop")
ax.scatter3D(x3_3, y3_3, 0.5*x3_3**2 + 2.5*y3_3**2, color='green', s=90, label="Adagrad")
ax.scatter3D(x3_4, y3_4, 0.5*x3_4**2 + 2.5*y3_4**2, color='blue', s=90, label="Adam")

ax.view_init(30, 15)
plt.legend(fontsize=20)
plt.show()
```
