```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
```

$y = 2 + 5x + 7x^{2} + 8x^{3} + \varepsilon$, for $x$ in $(-2, 2)$

```python
def f(x):
    return 2 + 5*x + 7*x**2 + 8*x**3
```

```python
x = np.linspace(-2, 2, 200)
fun_data = np.array([f(i) for i in x])
epsilon = np.random.normal(0, 0.2, 200)
y = fun_data + epsilon
```

```python
x_train = x[0:160].reshape(160, 1)
y_train = y[0:160].reshape(160, 1)
x_pre = x[160:].reshape(40, 1)
y_pre = y[160:].reshape(40, 1)
```

## Degree 2 :

### step 1.

```python
H2_xtrain = np.hstack([np.power(x_train, i) for i in range(0, 3)])
H2_xtest = np.hstack([np.power(x_pre, i) for i in range(0, 3)])
print("H(Xtrain) = \n", H2_xtrain)
```

### step 2.

```python
B2 = np.linalg.inv(np.dot(H2_xtrain.T, H2_xtrain))
A2 = np.dot(np.dot(B2, H2_xtrain.T), y_train)
print("A = \n", A2)
```

### step 3.

```python
f2_xtest = np.dot(H2_xtest, A2)
f2_xtrain = np.dot(H2_xtrain, A2)
print("f(Xtrain) = \n", f2_xtrain)
```

### step 4.

```python
# error between Ytrain and f(Xtrain) : 
Error_train_2 = np.sqrt(np.dot((f2_xtrain - y_train).T, (f2_xtrain - y_train)))
print("Order 2, train error : \n", Error_train_2)
```

### step 5.

```python
# error between Ypre and f(Xtest) : 
Error_test_2 = np.sqrt(np.dot((f2_xtest - y_pre).T, (f2_xtest - y_pre)))
print("Order 2, test error : \n", Error_test_2)
```
