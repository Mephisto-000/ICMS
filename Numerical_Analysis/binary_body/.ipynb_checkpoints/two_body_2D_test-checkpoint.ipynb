## ODE Model : 

$
\begin{align}
\frac{d\hat{r}_{i}}{dt}  = & \hat{v}_{i} \\
\frac{d\hat{v}_{i}}{dt}  = & m_{j}\frac{G\hat{r}_{ij}}{r_{ij^{3}}} \\
\end{align}
$

$G = 6.672\cdot 10^{-11} \text{ }Nm^{2}/kg^{2}$

$I.C. : $

$
\begin{align}
M_{A} =& 1.989\cdot 10^{30}\cdot (1.1), \text{ }kg  \\
M_{B} =& 1.989\cdot 10^{30}\cdot (0.907), \text{ }kg  \\
\hat{r}_{A,0} =& (0, 0, 0), \text{ }(m)  \\
\hat{r}_{B,0} =& (23AU, 0, 0), \text{ }(m)  \\
\hat{v}_{A,0} =& (0, 2000, 0), \text{ }(m/s.)  \\
\hat{v}_{B,0} =& (0, -1800, 0), \text{ }(m/s.),  \\
\end{align}
$

$AU = 1.496\cdot 10^{11} \text{ }m$

## Plot : 

```python
AU = 1.496e11
sec_in_day = 24 * 3600
t = 36500 * sec_in_day  # 100 years
dt = 100 * sec_in_day   # per 100 days
```

```python
alpha_centauri_A = Star("Alpha Centauri A", [0.0, 0.0, 0.0], [0.0, 2234.5, 0.0], 1.1*1.989e30)
alpha_centauri_B = Star("Alpha Centauri B", [23*AU, 0.0, 0.0], [0.0, -1810, 0.0], 0.907*1.989e30)
```

```python
Test = Simulation(stars=[alpha_centauri_A, alpha_centauri_B], t=t, dt=dt)
Test.shift_pos_vel()
```

```python
Test.plot_result()
```
